# Basic Mathematics (arithmetic, trigonometry, etc.)

```python
import numpy as np
```

Adição, subração, divisão e mutiplicação

```python
a = np.array([1,2,3])
a + 10 # Adiciona 10 a cada elemento
```

```python
a = np.array([1,2,3])
a - 10 # Subtrai 10 de cada elemento
```

```python
a = np.array([1,2,3])
a * 10 # multiplica por 10 cada elemento
```

```python
a = np.array([1,2,3])
a / 10 # Divide por 10 cada elemento
```

Oparações com mais de um array

```python
a = np.array([1,2,3,4,5])
b = np.array([5,4,3,2,1])

a+b # soma cada elemento com seu índice equivalente
```

```python
a = np.array([1,2,3,4,5])
b = np.array([5,4,3,2,1])

a-b # reduz cada elemento por seu índice equivalente
```

```python
a = np.array([1,2,3,4,5])
b = np.array([5,4,3,2,1])

a*b # multiplica cada elemento por seu índice equivalente
```

```python
a = np.array([1,2,3,4,5])
b = np.array([5,4,3,2,1])

a/b # divide cada elemento por seu índice equivalente
```

potencias

```python
a = np.array([1,2,3])
a ** 2 # eleva todos ao quadrado
```

Seno

```python
a = np.array([0, np.pi/2, 5, 8])
np.sin(a)# Pega o seno de cada número
```

Cosseno

```python
a = np.array([0, np.pi/2, 5, 8])
np.cos(a)# Pega o cosseno de cada número
```
