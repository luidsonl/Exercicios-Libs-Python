# Linear Algebra

```python
import numpy as np
```

Multiplicando uma matriz por outra

```python
a = np.full((2,3),2)
b = np.full((3,2),3)

np.matmul(a,b)
```

Encontrando o determinante da matriz

```python
c = np.identity(3)
np.linalg.det(c)
```

```python
c = np.array([[1,2,3],[1,2,3],[1,2,3]])
np.linalg.det(c)
```

```python
c = np.array([[5,2,2],[7,4,9],[4,6,3]])
np.linalg.det(c)
```
