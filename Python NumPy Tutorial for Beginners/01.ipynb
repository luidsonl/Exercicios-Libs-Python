# The Basics (creating arrays, shape, size, data type)

```python
# importando numpy
import numpy as np
```

Criando array unidimencional

```python
a = np.array([1,2,3,4,5,6,7])
print(a)
```

Criando array com duas dimensões

```python
b = np.array([[1,2,3,4,5,6,7],[7,6,5,4,3,2,1]])
print(b)
```

Pegando o número de dimenções

```python
# a
a.ndim
```

```python
# b
b.ndim
```

Pegando o shape

```python
# a
a.shape
```

```python
# b
b.shape
```

Pegando o type

```python
a.dtype
```

```python
b.dtype
```

Pegando o size

```python
a.itemsize
```

```python
b.itemsize
```

## Definindo o dataType

```python
a = np.array([1,2,3,4,5,6,7], dtype='int16')
b = np.array([[1,2,3,4,5,6,7],[7,6,5,4,3,2,1]],dtype='int32')

print(a.itemsize)
print(b.itemsize)
```

## Pegando o tamanho total

```python
a.nbytes
```

```python
b.nbytes
```
