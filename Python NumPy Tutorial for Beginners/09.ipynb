# Reorganizing Arrays (reshape, vstack, hstack)

```python
import numpy as np
```

```python
antes = np.array([[1,2,3,4],[5,6,7,8]])
antes.shape # informa que exitem 2 linhas e 4 colunas
```

```python
depois = antes.reshape(1,8) # coloca todos em uma linha
depois
```

```python
depois = antes.reshape(4,2)
depois
```

```python
depois = antes.reshape(2,2,2) # cria uma matriz tridimensional
depois
```

Empilhamento vertical

```python
l1 = np.array([1,2,3,4])
l2 = np.array([5,6,7,8])

np.vstack([l1,l1,l2,l2])
```

Empilhamento horizontal

```python
c1 = np.array([[1],[2],[3]])
c2 = np.array([[4],[5],[6]])

np.hstack([c1,c2])
```
