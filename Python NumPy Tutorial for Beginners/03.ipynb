# Initializing Different Arrays (1s, 0s, full, random, etc...)

```python
import numpy as np
```

Criando um array de zeros

```python
np.zeros(3)
```

```python
np.zeros((3,3))
# gera um array de 2 dimensões
```

```python
np.zeros((3,3,3))
# gera um array de 3 dimensões
```

```python
np.zeros((3,3,3,3))
# gera um array de 4 dimensões
```

array com 1s

```python
np.ones((3,3),dtype='int16')
# também é possível especificar o datatype
```

Criando array formado apenas por um número específico

```python
np.full((3,3),6)
```

Criando a partir de um shape específico

```python
a = np.array([[[111,112],[121,122]],[[211,212],[221,222]]])
#array 2x2x2
print(a)
```

```python
np.full_like(a,3)
```

Números decimais randômicos

```python
np.random.rand(2,2)
```

```python
np.random.rand(2,2,3)
```

```python
# pegando o shape
np.random.random_sample(a.shape)
```

Random de inteiros

```python
np.random.randint(10,20) #retorna números entre 10 e 29
```

```python
np.random.randint(1,20,size=(3,3)) # gera um array 3x3
```

Criando matriz identidade

```python
np.identity(6)
```

Repetindo arrays

```python
arr = np.array([[1,2,3,4]])# criando array bidimensional
r1 = np.repeat(arr,4, axis=0) #vai repetir o array 4 vezes no eixo 0
print(r1)
```

```python
r2 = np.repeat(arr,4, axis=1) #vai repetir o array 4 vezes no eixo 1
print(r2)
```
