# Statistics

```python
import numpy as np
```

```python
a = np.array([[1,2,3],[4,5,6]])
a
```

```python
np.min(a) #Vai retonar o valor mínimo
```

```python
np.min(a, axis=1) #Vai retonar o valor mínimo de cada linha
```

```python
np.max(a) #Vai retonar o valor máximo
```

```python
np.max(a, axis=0) #Vai retonar o valor máximo por coluna
```

```python
np.mean(a) #Vai retonar a média
```

```python
np.mean(a, axis=1) #Vai retonar a média de cada linha
```

```python
np.sum(a) # retorna a soma total
```
