# Be careful when copying variables!

```python
import numpy as np
```

```python
a = np.array([1,2,3])
b = a #'b' é uma cópia do endereço de memória de 'a'
print('a =',a)
print('b =',b)
```

```python
b[2] = -3
# alterando "a", "b" também é alterado
print('a =',a)
print('b =',b)
```

```python
a = np.array([1,2,3])
b = a.copy() #'b' não é uma cópia do mesmo endereço de memória de 'a'
print('a =',a)
print('b =',b)
```

```python
b[2] = -3

print('a =',a)
print('b =',b)
```
