### Quantidade de pokemons tipo água que tem mais de 100 de HP

```python
def filter_per_group_by_limit(df, col_group, col_confer, value):
    u_types = df[col_group].unique()
    df_list = {}
    
    for u_type in u_types:
        df_list[u_type] = df.loc[(df[col_group]== u_type) & (df[col_confer]>= value)]
        
    return df_list
```

```python
filter_per_group_by_limit(df, 'Type 1', 'HP', 100)['Water'].count()[0]
```

# Agregação com groupby

## Pegando a média dos grupos

```python
df2 = pd.DataFrame({'Animal': ['Falcon', 'Falcon','Parrot', 'Parrot'],
                   'Max Speed': [380., 370., 24., 26.]})

df2.groupby(['Animal']).mean()
```

```python
# Forma grupos a partir de valores iguais
#df.groupby(['Type 1']).mean() 
#dá erro por conta dos valores categóricos
df.select_dtypes(include='number').groupby(df['Type 1']).mean().sort_values('HP',ascending= False)
#pega apenas os valores numéricos
# ordena a partir da média de hp do maior para o menor
```

## Quantidade por grupo

```python
df.groupby(df['Type 1']).count().sort_values('#',ascending= False)['#']
```

```python
df.groupby(['Type 1', 'Type 2'])['#'].count()
```
