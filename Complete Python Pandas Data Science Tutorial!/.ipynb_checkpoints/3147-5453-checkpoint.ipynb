### Quantidade de pokemons tipo água que tem mais de 100 de HP

```python
def filter_per_group_by_limit(df, col_group, col_confer, value):
    u_types = df[col_group].unique()
    df_list = {}
    
    for u_type in u_types:
        df_list[u_type] = df.loc[(df[col_group]== u_type) & (df[col_confer]>= value)]
        
    return df_list
```

```python
filter_per_group_by_limit(df, 'Type 1', 'HP', 100)['Water'].count()[0]
```

# Regex

```python
df.loc[df['Name'].str.contains('Mega')].head()
#Pega todos os mega
```

```python
df.loc[~df['Name'].str.contains('Mega')].head()
# Pega todos que não são mega
```

```python
import re
```

```python
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
# Pega todos que forem tipo fogo ou grama
```

```python
df.loc[df['Type 1'].str.contains('fIre|iCe',flags=re.I, regex=True)]
# Não case sensitive
```

```python
df.loc[df['Name'].str.contains('^Ra', regex=True)]
#Pega todos que o nome começa com 'Ra'
```
