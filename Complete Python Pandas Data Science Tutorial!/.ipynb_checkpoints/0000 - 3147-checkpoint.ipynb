## .loc

```python
df.loc[0,['Name']]
# Permite usar strings
```

```python
df.loc[df['Type 1']=='Water'].head(5)
```

## Exibindo dados com um loop

```python
for indice, linha in df.iloc[0:9].iterrows():
    print(indice, linha['Name'])
```

```python
df.iloc[0:9]['Name']
# Quase a mesma coisa
```

# Sort/Describe

```python
df.describe()
```

```python
df.sort_values('Name')
# Ordena pelos nomes em ordem alfabética
```

```python
df.sort_values('Speed', ascending=False)
# Ordena do mais rápido para o mais lento
```

```python
df.sort_values(['Speed', 'Attack'], ascending=[0,1])
# Caso os valores sejam iguais, ele irá levar em consideração o segindo parâmetro do array
```

# Fazendo mudanças nos dados

## Adicionando nova coluna

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.columns
```

```python
df.head()
```

## Removendo coluna

```python
df.drop(columns = ['Total'], inplace=True)
```

```python
df
```

## Adicionando usando .iloc

```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df
```

# Rearranjando colunas

```python
cols = list(df.columns.values)
df = df[cols[0:10] + [cols[-1]] + cols[10:-1]]
df
```

# Salvando dados

```python
df.to_csv('data/modified/pokemon.csv', index = False)
```

```python
df.to_excel('data/modified/pokemon.xlsx', index = False)
```

```python
df.to_csv('data//modified/pokemon.txt', index = False, sep = '\t')
```
